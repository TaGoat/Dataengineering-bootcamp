```python
import numpy as np

def vectorize(words, vocabulary):
    vector = np.zeros(len(vocabulary))
    for i, word in enumerate(vocabulary):
        if word in words:
            vector[i] = 1
    return vector


text1 = "i love tareq hello word"
text2 = "tareq love coding hello world"

words1 = text1.lower().split()
words2 = text2.lower().split()
vocabulary = sorted(set(words1 + words2))

vector1 = vectorize(words1, vocabulary)
vector2 = vectorize(words2, vocabulary)

print("Vocabulary:", vocabulary)
print("Vector 1:", vector1)
print("Vector 2:", vector2)
```
