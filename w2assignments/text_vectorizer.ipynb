```python
def vectorize_texts(text1, text2):
    words1 = text1.lower().split()
    words2 = text2.lower().split()
    all_words = words1 + words2
    unique_words = sorted(set(all_words))
    vector1 = []
    vector2 = []
    for word in unique_words:
        vector1.append(words1.count(word))
        vector2.append(words2.count(word))
    return vector1, vector2, unique_words


text1 = "i love tareq hello word"
text2 = "tareq love coding hello world"

vector1, vector2, unique_words = vectorize_texts(text1, text2)

print("Vocabulary:", unique_words)
print("Vector 1:", vector1)
print("Vector 2:", vector2)
```
